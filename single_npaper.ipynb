```python
# Printing means, medians and STDs
print('Fav_count mean: ', tweet_pd_filtered['fav_count'].mean())
print('Fav_count median: ', tweet_pd_filtered['fav_count'].median())
print('Retw_count mean: ', tweet_pd_filtered['retw_count'].mean())
print('Retw_count median: ', tweet_pd_filtered['retw_count'].median())
print('Fav_count standard deviation: ', tweet_pd_filtered['fav_count'].std())
print('Retw_count standard deviation: ', tweet_pd_filtered['retw_count'].std())
```
